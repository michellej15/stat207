### 4. Select a categorical variable from this dataset and create a barplot that shows the *counts* for each level of this categorical variable. [1 pt]

```python
citric_acid = df["citric acid"].value_counts()
sns.barplot(x = citric_acid.index, y = citric_acid)
plt.title("Citric Acid Content in Red and White Portuguese Vinho Verde wine")
plt.ylabel("Count of Citric Acid")
plt.xlabel("Citric Acid Content")
sns.set(rc={'figure.figsize':(11,20)})
plt.show()
```
